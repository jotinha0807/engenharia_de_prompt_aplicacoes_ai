```markdown
# Organizador de Arquivos por Extensão

Este programa organiza arquivos em um diretório específico, movendo-os para subpastas que levam o nome de suas respectivas extensões. Isso é útil para manter a ordem em pastas que contêm muitos tipos de arquivos diferentes.
```

```markdown
## Passo 1: Preparar o Ambiente

Primeiro, vamos importar as bibliotecas necessárias e criar um diretório para nossos arquivos de teste.
```

````python
```python
import os
import shutil

# Definir o diretório onde os arquivos bagunçados serão criados e organizados
# É uma boa prática criar um diretório de teste para não bagunçar seus arquivos reais.
# Vamos usar 'arquivos_para_organizar' como nosso diretório de trabalho.
 disorganized_folder = 'arquivos_para_organizar'

# Criar o diretório se ele não existir
if not os.path.exists(disorganized_folder):
    os.makedirs(disorganized_folder)
    print(f"Diretório '{disorganized_folder}' criado com sucesso.")
else:
    print(f"Diretório '{disorganized_folder}' já existe.")
```
````

```markdown
## Passo 2: Gerar Arquivos de Exemplo

Para demonstrar o programa, vamos criar 10 arquivos de cada tipo (pdf, jpg, xls, doc, ipynb) dentro da pasta `arquivos_para_organizar`.
```

````python
```python
# Lista de extensões que vamos usar
extensions = ['pdf', 'jpg', 'xls', 'doc', 'ipynb']
num_files_per_type = 10

print(f"Gerando {num_files_per_type} arquivos para cada extensão em '{disorganized_folder}'...")

for ext in extensions:
    for i in range(1, num_files_per_type + 1):
        file_name = f"arquivo_exemplo_{ext}_{i}.{ext}"
        file_path = os.path.join(disorganized_folder, file_name)
        # Criar um arquivo vazio (ou com conteúdo mínimo) para cada extensão
        with open(file_path, 'w') as f:
            f.write(f"Este é o conteúdo do arquivo {file_name}\n")

print("Arquivos de exemplo gerados.")

# Listar os arquivos criados para verificar
print("Arquivos na pasta antes da organização:")
for f in os.listdir(disorganized_folder):
    print(f"- {f}")
```
````

```markdown
## Passo 3: Implementar a Lógica de Organização

Agora, vamos escrever a função principal que percorre os arquivos, identifica suas extensões, cria as pastas correspondentes e move os arquivos.
```

````python
```python
def organize_files_by_extension(directory):
    """
    Organiza arquivos em um diretório movendo-os para subpastas
    baseadas em suas extensões.

    Args:
        directory (str): O caminho para o diretório a ser organizado.
    """
    print(f"\nIniciando a organização de arquivos no diretório: '{directory}'")

    # Listar todos os itens no diretório
    for filename in os.listdir(directory):
        # Construir o caminho completo do arquivo
        file_path = os.path.join(directory, filename)

        # Verificar se é um arquivo (e não um subdiretório já existente)
        if os.path.isfile(file_path):
            # Separar o nome do arquivo da sua extensão
            # os.path.splitext retorna uma tupla: ('nome_do_arquivo', '.extensao')
            _, extension = os.path.splitext(filename)

            # Remover o ponto inicial da extensão e convertê-la para minúsculas
            # Ex: '.pdf' se torna 'pdf'
            if extension:
                extension = extension[1:].lower()
            else:
                # Se não houver extensão (ex: 'Makefile'), podemos classificá-lo como 'sem_extensao'
                extension = 'sem_extensao'

            # Definir o nome da pasta de destino para esta extensão
            destination_folder = os.path.join(directory, extension)

            # Criar a pasta de destino se ela não existir
            if not os.path.exists(destination_folder):
                os.makedirs(destination_folder)
                print(f"  Pasta '{extension}/' criada.")

            # Mover o arquivo para a pasta de destino
            shutil.move(file_path, destination_folder)
            print(f"  Movido '{filename}' para '{extension}/'.")

    print(f"\nOrganização concluída para o diretório: '{directory}'.")
```
````

```markdown
## Passo 4: Executar a Organização

Agora vamos chamar a função com o nosso diretório de exemplo.
```

````python
```python
# Chamar a função para organizar os arquivos
organize_files_by_extension(disorganized_folder)

print("\nVerificando o conteúdo da pasta após a organização:")
for item in os.listdir(disorganized_folder):
    item_path = os.path.join(disorganized_folder, item)
    if os.path.isdir(item_path):
        print(f"- Pasta '{item}/':")
        for sub_item in os.listdir(item_path):
            print(f"  - {sub_item}")
    else:
        print(f"- Arquivo restante: {item}") # Não deve haver arquivos aqui após a organização
```
````
