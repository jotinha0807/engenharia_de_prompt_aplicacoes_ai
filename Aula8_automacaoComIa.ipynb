```python
def verificar_temperatura():
    """Verifica a temperatura do sistema e exibe um alerta se estiver acima de 30 graus."""
    try:
        temperatura = float(input("Digite a temperatura atual do sistema (em Celsius): "))

        if temperatura > 30:
            print(f"ALERTA: Superaquecimento! Temperatura atual: {temperatura}°C")
        else:
            print(f"Temperatura normal. Temperatura atual: {temperatura}°C")
    except ValueError:
        print("Entrada inválida. Por favor, digite um número para a temperatura.")

# Chamada da função para executar a verificação
verificar_temperatura()
```
